# Getting Files From Drive

```python
train_cat = pd.read_csv('/content/gdrive/My Drive/Research Paper/Data/category_balanced.csv')
val_cat = pd.read_csv('/content/gdrive/My Drive/Research Paper/Data/test/category/category_train_val.csv')
test_cat = pd.read_csv('/content/gdrive/My Drive/Research Paper/Data/test/category/category_test.csv')
```

```python
train_sat = pd.read_csv('/content/gdrive/My Drive/Research Paper/Data/satisfaction_balanced.csv')
val_sat = pd.read_csv('/content/gdrive/My Drive/Research Paper/Data/test/satisfaction/satisfaction_train_val.csv')
test_sat = pd.read_csv('/content/gdrive/My Drive/Research Paper/Data/test/satisfaction/satisfaction_test.csv')
```

```python
train_env = pd.read_csv('/content/gdrive/My Drive/Research Paper/Data/environment_balanced.csv')
val_env = pd.read_csv('/content/gdrive/My Drive/Research Paper/Data/test/environment/environment_train_val.csv')
test_env = pd.read_csv('/content/gdrive/My Drive/Research Paper/Data/test/environment/environment_test.csv')
```

# Splitting Files

```python
X_train_cat = train_cat['text'].to_list()
y_train_cat = train_cat['category'].to_list()

X_val_cat = val_cat['text'].to_list()
y_val_cat = val_cat['category'].to_list()

X_test_cat = test_cat['text'].to_list()
y_test_cat = test_cat['category'].to_list()
```

```python
X_train_sat = train_sat['text'].to_list()
y_train_sat = train_sat['satisfaction'].to_list()

X_val_sat = val_sat['text'].to_list()
y_val_sat = val_sat['satisfaction'].to_list()

X_test_sat = test_sat['text'].to_list()
y_test_sat = test_sat['satisfaction'].to_list()
```

```python
X_train_env = train_env['text'].to_list()
y_train_env = train_env['environment'].to_list()

X_val_env = val_env['text'].to_list()
y_val_env = val_env['environment'].to_list()

X_test_env = test_env['text'].to_list()
y_test_env = test_env['environment'].to_list()
```

# Implementing Model

```python
label_list = ['Autonomy', 'Competence', 'Relatedness']

MODEL_NAME = 'distilbert-base-uncased'
t_cat = text.Transformer(MODEL_NAME, maxlen=500,
                     classes=label_list)


# trn_cat, val_cat, preproc_cat = text.texts_from_array(x_train=X_train_cat, y_train=y_train_cat,
#                                           x_test=X_val_cat, y_test=y_val_cat,
#                                           class_names=label_list,
#                                           preprocess_mode='distilbert',
#                                           maxlen=350)

# model_cat = text.text_classifier('distilbert', train_data=trn_cat, preproc=preproc_cat)
# learner_cat = ktrain.get_learner(model_cat, train_data=trn_cat, val_data=val_cat, batch_size=6)

train_cat = t_cat.preprocess_train(X_train_cat, y_train_cat)
val_cat = t_cat.preprocess_test(X_val_cat, y_val_cat)

model_cat = t_cat.get_classifier()
learner_cat = ktrain.get_learner(model_cat, train_data=train_cat, val_data=val_cat, batch_size=6)

learner_cat.fit_onecycle(3e-5, 3)
```

```python
label_list = ['Satisfied', 'Dissatisfied', 'Neutral', 'Not Clear']

MODEL_NAME = 'distilbert-base-uncased'
t_sat = text.Transformer(MODEL_NAME, maxlen=500,
                     classes=label_list)


# trn_sat, val_sat, preproc_sat = text.texts_from_array(x_train=X_train_sat, y_train=y_train_sat,
#                                           x_test=X_val_sat, y_test=y_val_sat,
#                                           class_names=label_list,
#                                           preprocess_mode='distilbert',
#                                           maxlen=350)

# model_sat = text.text_classifier('distilbert', train_data=trn_sat, preproc=preproc_sat)
# learner_sat = ktrain.get_learner(model_sat, train_data=trn_sat, val_data=val_sat, batch_size=6)

train_sat = t_sat.preprocess_train(X_train_sat, y_train_sat)
val_sat = t_sat.preprocess_test(X_val_sat, y_val_sat)

model_sat = t_sat.get_classifier()
learner_sat = ktrain.get_learner(model_sat, train_data=train_sat, val_data=val_sat, batch_size=6)

learner_sat.fit_onecycle(3e-5, 3)
```

```python
label_list = ['Supportive', 'Non-Supportive', 'Not Clear']

MODEL_NAME = 'distilbert-base-uncased'
t_env = text.Transformer(MODEL_NAME, maxlen=500,
                     classes=label_list)

# trn_env, val_env, preproc_env = text.texts_from_array(x_train=X_train_env, y_train=y_train_env,
#                                           x_test=X_val_env, y_test=y_val_env,
#                                           class_names=label_list,
#                                           preprocess_mode='distilbert',
#                                           maxlen=350)

# model_env = text.text_classifier('distilbert', train_data=trn_env, preproc=preproc_env)
# learner_env = ktrain.get_learner(model_env, train_data=trn_env, val_data=val_env, batch_size=6)

train_env = t_env.preprocess_train(X_train_env, y_train_env)
val_env = t_env.preprocess_test(X_val_env, y_val_env)

model_env = t_env.get_classifier()
learner_env = ktrain.get_learner(model_env, train_data=train_env, val_data=val_env, batch_size=6)

learner_env.fit_onecycle(3e-5, 3)
```
